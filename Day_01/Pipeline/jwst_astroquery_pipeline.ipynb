## Astroquery for JWST

```python
from astroquery.mast import Observations
from astropy.table import vstack
```

```python

obs_table = Observations.query_criteria(proposal_id="1727",
                                        instrument_name="NIRCAM*",
                                        filters="F444W",
                                        obs_collection="JWST")
```

```python
len(obs_table)
```

```python
obs_table[0]
```

```python
# Query each observation in a loop to avoid server-side table joins and timeouts
product_list = [Observations.get_product_list(obs) for obs in obs_table]
products = vstack(product_list)
```

```python
len(products)
```

```python
mask = ["uncal" in row["dataURI"] for row in products]
uncal_products = products[mask]
len(uncal_products)
```

```python
uncal_products
mask2 = ["jw01727045" in row["dataURI"] for row in uncal_products]
uncal45_products = uncal_products[mask2]
len(uncal45_products)
```

## Run the pipeline

```python
# Setup CRDS
import os


os.environ["CRDS_SERVER_URL"] = "https://jwst-crds.stsci.edu"
os.environ["CRDS_PATH"] = "/media/home/data/jwst/crds/crds_cache"
```

```python
from multiprocessing import Pool
from glob import glob

from jwst.pipeline import Detector1Pipeline, Image2Pipeline, Image3Pipeline
from threadpoolctl import threadpool_limits
```

```python
uncal_files = glob("*_uncal.fits")
uncal_files.sort()
uncal_files
```

```python
@threadpool_limits.wrap(limits=1, user_api='blas')
def rundetector1(file):
    steps = {
        "jump": {
            "save_results": True,
        },
    }
    result = Detector1Pipeline.call(file, save_results=True, steps=steps)

with Pool(processes=4) as pool:
    pool.map(rundetector1, uncal_files)
```

```python
rate_files = glob("*_rate.fits")
rate_files.sort()
rate_files
```

```python
for file in rate_files:
    result = Image2Pipeline.call(file, save_results=True)
```

```python
result = Image3Pipeline.call("jw01727-o045_20230718t020758_image3_00003_asn.json", save_results=True)
```
